```python

test_ds_path = ".\..\Datasets\store_data.csv"

with open(test_ds_path, "r") as f:

    lines = f.readlines()
    inventory = list()
    tuples = list()

    for line in lines:
        transaction = line.strip().split(",")
        #print(transaction)
        tuples.append(transaction)
        for element in transaction:
            if element.strip().replace(' ','_').replace('&','and') not in inventory:
                inventory.append(element.strip().replace(' ','_').replace('&','and'))
```

```python
import mask
```

```python
print(inventory)
inventory.pop()
inventory.sort()
print(inventory)
```

```python
import pandas as pd
store_ds = pd.DataFrame(
    [[1 if item in row else 0 for item in inventory] for row in tuples],
    columns=inventory
)
```

```python
from mask.functions import MASK_frequent_itemsets,MASK_Distortion
from mask.eval import support_error,identity_error,P
from mask.utility.utility_fun import mean_support,Apriori
```

```python
mean_sup = mean_support(inventory,store_ds)
print(mean_sup)
```

```python
p = 0.9
min_sup = 0.0025
```

```python
distorted_ds = MASK_Distortion(store_ds,p)
```

```python
frequent_itemsets = Apriori(store_ds,min_sup)
```

```python
#m_min_sup = min_sup - 0.25*min_sup
mask_frequent_intemset = MASK_frequent_itemsets(distorted_ds,p,min_sup)
```

```python
from decimal import *
getcontext().prec = 28
def support_error2(AprioriRuleslevel,MASKRuleslevel):
    
    sum = 0
    cnt = 0

    for apriorirule in AprioriRuleslevel:
        if apriorirule is not None and apriorirule in MASKRuleslevel:
            cnt += 1
            index = MASKRuleslevel.index(apriorirule)
            
            rec_sup = Decimal(MASKRuleslevel[index].support)
            act_sup = Decimal(apriorirule.support)
            
            sum += Decimal((abs(Decimal(rec_sup-act_sup))/act_sup))
    if cnt != 0:
        return (sum/cnt)*100
    else:
        return 0
```

```python
for i in range(1,len(frequent_itemsets)):
    print(f"Level: {i},|F| = {len(frequent_itemsets[i])}, Support error: {support_error2(frequent_itemsets[i],mask_frequent_intemset[i])}, Identity error: {identity_error(frequent_itemsets[i],mask_frequent_intemset[i])}")
```
